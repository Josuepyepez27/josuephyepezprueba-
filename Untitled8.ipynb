```python
import pandas as pd
import json
import os
dfs = []
folder_path = '/content'
for file_name in os.listdir(folder_path):
    if file_name.endswith('.json'):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r') as f:
            data = json.load(f)
            df = pd.json_normalize(data)
            dfs.append(df)
full_df = pd.concat(dfs, ignore_index=True)
full_df.head()
```

```python

print(f"Cantidad de registros cargados: {len(full_df)}")
```

```python

tipo_centros = full_df['centro_tipo_descripcion'].value_counts()
print(tipo_centros)
```

```python
distritos = full_df['distrito_nombre'].value_counts()
print(distritos)
import folium
m = folium.Map(location=[40.4168, -3.7038], zoom_start=12)
m
```

```python

m = folium.Map(location=[40.4168, -3.7038], zoom_start=12)
centros_publicos = full_df[full_df['centro_titularidad'] == 'PÚBLICO']
centros_privados = full_df[full_df['centro_titularidad'] == 'PRIVADO']
for _, row in centros_privados.iterrows():
    folium.Marker([row['direccion_coor_y'], row['direccion_coor_x']],
                  popup=row['centro_nombre'], icon=folium.Icon(color='red')).add_to(m)
m
```

```python

distribucion_distrito_tipo = full_df.groupby(['distrito_nombre', 'centro_tipo_descripcion']).size().unstack()
print(distribucion_distrito_tipo)
```

```python
centros_por_cp = full_df.groupby('direccion_codigo_postal').size().mean()
print(f"Promedio de centros educativos por código postal: {centros_por_cp}")
```
